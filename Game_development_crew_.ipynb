````python
%%writefile main.py
### Paste your code below
import pygame, sys, random

# Instructions for playing the game
#
# **How to Play Sir Hopsalot's Wobbly Jaunt:**
# 1. Run the script: `python your_game_file.py`
# 2. Controls:
#    - Press **SPACEBAR** or the **UP ARROW** key to make Sir Hopsalot jump.
#      He has a fixed jump height and cannot double-jump.
#    - Press **ESCAPE** to quit the game at any time.
# 3. Objective: Guide Sir Hopsalot over and across an endless stream of obstacles.
# 4. Scoring:
#    - Your score increases simply by surviving.
#    - You get bonus points for successfully clearing obstacles.
# 5. Game Over:
#    - Collide with any obstacle (walls, gnomes, fowl) or fall into a puddle, and the game ends.
#    - On the "Game Over!" screen, press **SPACEBAR** to restart.
#
# Have a wobbly good time!

# Initialize Pygame and its mixer for sounds
pygame.init()
pygame.mixer.init()

# Screen dimensions
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 400
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("Pulse Jump")

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
GREEN = (100, 200, 100)
BLUE_SKY = (135, 206, 235)
BROWN = (139, 69, 19)
RED = (255, 0, 0)
GREY = (150, 150, 150)
YELLOW = (255, 255, 0)
DARK_GREEN = (50, 150, 50) # For gnome body and ground texture

# Game variables
GROUND_LEVEL = 350
GRAVITY = 0.6
GAME_SPEED = 5 # Speed at which environment scrolls
raw_score = 0 # Internal score, increments every frame
GAME_ACTIVE = True
FONT = pygame.font.Font(None, 36)
LARGE_FONT = pygame.font.Font(None, 72)

# Obstacle spawning constants
OBSTACLE_SPAWN_TIMER = 0
OBSTACLE_SPAWN_INTERVAL_MIN = 60 # frames (1 second at 60 FPS)
OBSTACLE_SPAWN_INTERVAL_MAX = 120 # frames (2 seconds at 60 FPS)
LAST_OBSTACLE_X = SCREEN_WIDTH # Tracks the rightmost edge of the last spawned obstacle

# Score thresholds for new obstacles
SCORE_THRESHOLD_GNOME = 100 # Display score points (raw_score // 10)
SCORE_THRESHOLD_FOWL = 250  # Display score points (raw_score // 10)

# Sound placeholders (actual sound files are not provided in this text-based output)
# In a real game, you would load sound files like:
# jump_sound = pygame.mixer.Sound('data/boing.wav')
# collision_sound = pygame.mixer.Sound('data/oof.wav')
def play_jump_sound():
    # If sound files were available: jump_sound.play()
    pass

def play_collision_sound():
    # If sound files were available: collision_sound.play()
    pass

# Player class: Sir Hopsalot, the main character
class SirHopsalot(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        # Initial dimensions for the player sprite
        self.original_width = 40
        self.original_height = 40
        # Create a transparent surface for custom drawing. This surface
        # and its rect are primarily used for collision detection and
        # to correctly size the custom drawn shapes.
        self.image = pygame.Surface([self.original_width, self.original_height], pygame.SRCALPHA)
        # Position the player on the ground
        self.rect = self.image.get_rect(center=(100, GROUND_LEVEL - self.original_height / 2))
        self.y_velocity = 0
        self.on_ground = True
        self.jump_strength = -12 # How high Sir Hopsalot jumps

        self.body_color = WHITE
        self.eye_color = BLACK
        self.foot_color = BROWN

        self.squash_stretch_factor = 0 # Factor for visual squash/stretch effect

    # Initiates a jump if Sir Hopsalot is on the ground
    def jump(self):
        if self.on_ground:
            self.y_velocity = self.jump_strength
            self.on_ground = False
            play_jump_sound()
            self.squash_stretch_factor = -0.1 # Slight squash at the start of a jump

    # Applies gravity to Sir Hopsalot's vertical movement
    def apply_gravity(self):
        self.y_velocity += GRAVITY
        self.rect.y += self.y_velocity

        # Prevent falling through the ground
        if self.rect.bottom >= GROUND_LEVEL:
            self.rect.bottom = GROUND_LEVEL
            self.y_velocity = 0
            self.on_ground = True
            # Apply stretch only if not already animating
            if abs(self.squash_stretch_factor) < 0.01:
                self.squash_stretch_factor = 0.1 # Slight stretch on landing

    # Updates the player's state, including gravity and squash/stretch animation
    def update(self):
        self.apply_gravity()

        # Animate squash/stretch
        target_width = self.original_width
        target_height = self.original_height

        if abs(self.squash_stretch_factor) > 0.01: # Continue animation if factor is significant
            target_width = int(self.original_width * (1 + self.squash_stretch_factor))
            target_height = int(self.original_height * (1 - self.squash_stretch_factor))
            # Ease the factor back to normal
            self.squash_stretch_factor *= 0.85
            if abs(self.squash_stretch_factor) < 0.01: # Snap to 0 when almost done
                self.squash_stretch_factor = 0

        # Update the surface for drawing (important for custom-drawn sprites).
        # We recreate `self.image` and `self.rect` to reflect the squash/stretch dimensions.
        # The custom `draw` method will use `self.rect` for its drawing calculations.
        old_center = self.rect.center
        self.image = pygame.Surface([target_width, target_height], pygame.SRCALPHA)
        self.rect = self.image.get_rect(center=old_center)

    # Draws Sir Hopsalot on the screen with his body, eyes, and feet
    def draw(self, screen_surface):
        # Body (ellipse to represent rotundness, adapts to squash/stretch via self.rect)
        pygame.draw.ellipse(screen_surface, self.body_color, self.rect)

        # Eyes (two smaller circles, positioned relative to the body's current rect)
        eye_radius = max(2, self.rect.width // 8) # Ensure radius is at least 2
        eye_offset_x = self.rect.width // 6
        eye_offset_y = self.rect.height // 6

        pygame.draw.circle(screen_surface, self.eye_color,
                           (self.rect.centerx - eye_offset_x, self.rect.centery - eye_offset_y),
                           eye_radius)
        pygame.draw.circle(screen_surface, self.eye_color,
                           (self.rect.centerx + eye_offset_x, self.rect.centery - eye_offset_y),
                           eye_radius)

        # Feet (two tiny circles, positioned at the bottom of the body)
        foot_radius = max(1, self.rect.width // 10) # Ensure radius is at least 1
        foot_offset_x = self.rect.width // 5
        foot_offset_y = self.rect.height // 8 # Closer to the bottom

        pygame.draw.circle(screen_surface, self.foot_color,
                           (self.rect.centerx - foot_offset_x, self.rect.bottom - foot_offset_y),
                           foot_radius)
        pygame.draw.circle(screen_surface, self.foot_color,
                           (self.rect.centerx + foot_offset_x, self.rect.bottom - foot_offset_y),
                           foot_radius)

# Obstacle class: Base for all game obstacles
class Obstacle(pygame.sprite.Sprite):
    def __init__(self, x, y, width, height, obstacle_type, color=GREY):
        super().__init__()
        self.image = pygame.Surface([width, height], pygame.SRCALPHA) # Use SRCALPHA for transparency
        self.image.fill(color)
        self.rect = self.image.get_rect(bottomleft=(x, y))
        self.type = obstacle_type
        self.cleared = False # Flag for scoring when player passes it

    # Updates the obstacle's position (moves left)
    def update(self):
        self.rect.x -= GAME_SPEED
        # Implement subtle wobbly animation for "wall" type
        if self.type == "wall":
            # Very small, infrequent random vertical shift
            if random.random() < 0.02: # 2% chance per frame to wobble
                self.rect.y += random.choice([-1, 1])
                # Ensure it doesn't wobble too far off the ground
                if self.rect.bottom > GROUND_LEVEL: self.rect.bottom = GROUND_LEVEL
                # Prevent walls from 'floating' too high
                if self.rect.top < GROUND_LEVEL - 100: self.rect.top = GROUND_LEVEL - 100

    # Draws the obstacle on the screen, with special visuals for different types
    def draw(self, screen_surface):
        screen_surface.blit(self.image, self.rect)
        if self.type == "gnome":
            # Simple gnome visual elements: hat, face
            hat_color = RED
            face_color = BROWN
            # Hat (triangle)
            pygame.draw.polygon(screen_surface, hat_color, [
                (self.rect.centerx, self.rect.top - 10), # Top point of hat
                (self.rect.left + 5, self.rect.centery - 10), # Left base of hat
                (self.rect.right - 5, self.rect.centery - 10) # Right base of hat
            ])
            # Face (oval)
            pygame.draw.ellipse(screen_surface, face_color,
                                pygame.Rect(self.rect.centerx - 8, self.rect.centery - 15, 16, 20))
            # Eyes
            pygame.draw.circle(screen_surface, BLACK, (self.rect.centerx - 4, self.rect.centery - 10), 2)
            pygame.draw.circle(screen_surface, BLACK, (self.rect.centerx + 4, self.rect.centery - 10), 2)
        elif self.type == "fowl":
            # Simple floating fowl visual elements: body, eye, beak
            fowl_color = YELLOW
            beak_color = RED
            # Body (main circle)
            pygame.draw.circle(screen_surface, fowl_color, self.rect.center, self.rect.width // 2)
            # Eye
            pygame.draw.circle(screen_surface, BLACK, (self.rect.centerx + self.rect.width // 4, self.rect.centery - self.rect.height // 4), 2)
            # Beak (small triangle)
            pygame.draw.polygon(screen_surface, beak_color, [
                (self.rect.centerx + self.rect.width // 2, self.rect.centery - 5),
                (self.rect.centerx + self.rect.width // 2 + 10, self.rect.centery),
                (self.rect.centerx + self.rect.width // 2, self.rect.centery + 5)
            ])


# Function to reset all game variables for a new round
def reset_game():
    global raw_score, GAME_ACTIVE, obstacles, player, OBSTACLE_SPAWN_TIMER, LAST_OBSTACLE_X
    raw_score = 0
    GAME_ACTIVE = True
    player = SirHopsalot()
    obstacles.empty() # Clears all obstacles from the group
    OBSTACLE_SPAWN_TIMER = 0
    LAST_OBSTACLE_X = SCREEN_WIDTH # Reset last obstacle position for new spawning cycle

# Function to spawn new obstacles onto the screen
def spawn_obstacle():
    global LAST_OBSTACLE_X
    obstacle_type_choices = ["wall", "puddle"]
    # Introduce more challenging obstacles as the score increases based on display score
    if raw_score // 10 > SCORE_THRESHOLD_GNOME:
        obstacle_type_choices.append("gnome")
    if raw_score // 10 > SCORE_THRESHOLD_FOWL:
        obstacle_type_choices.append("fowl")

    selected_type = random.choice(obstacle_type_choices)

    # Calculate initial x position for the new obstacle, off-screen to the right
    x_offset = random.randint(50, 200)
    # Ensure there's a minimum clear distance from the previous obstacle
    min_clear_distance = player.original_width * 3 # At least 3 player widths gap
    x = max(SCREEN_WIDTH + x_offset, LAST_OBSTACLE_X + min_clear_distance)

    new_obstacle = None
    if selected_type == "wall":
        width = random.randint(30, 60)
        height = random.randint(40, 80)
        new_obstacle = Obstacle(x, GROUND_LEVEL, width, height, "wall", GREY)
    elif selected_type == "puddle":
        width = random.randint(60, 120)
        # Puddle's 'height' conceptually means its depth.
        # Its 'image' is just to represent the falling area.
        height = 10
        # Puddle's rect.top is set to GROUND_LEVEL for consistent collision logic,
        # visual depth is implicitly handled by the type.
        new_obstacle = Obstacle(x, GROUND_LEVEL + height, width, height, "puddle", BLUE_SKY)
        new_obstacle.rect.top = GROUND_LEVEL # The effective top for collision is at ground level
    elif selected_type == "gnome":
        width = 30
        height = 40
        new_obstacle = Obstacle(x, GROUND_LEVEL, width, height, "gnome", DARK_GREEN) # Gnome body color
    elif selected_type == "fowl":
        width = 30
        height = 25 # Smaller, floating obstacle
        # Floating fowl appears at a height roughly where player jumps
        y = random.randint(GROUND_LEVEL - player.original_height - 30, GROUND_LEVEL - player.original_height - 10)
        new_obstacle = Obstacle(x, y, width, height, "fowl", YELLOW) # Rubber chicken/duck color

    if new_obstacle:
        obstacles.add(new_obstacle)
        LAST_OBSTACLE_X = new_obstacle.rect.right # Update the last obstacle's right edge

# Game entities setup
player = SirHopsalot()
obstacles = pygame.sprite.Group() # Pygame group to manage all obstacles

# Ground surface for continuous scrolling
# Create a ground image that's twice the screen width for seamless scrolling
ground_image = pygame.Surface((SCREEN_WIDTH * 2, SCREEN_HEIGHT - GROUND_LEVEL))
ground_image.fill(GREEN)
# Draw some darker green patches on the ground for texture
for _ in range(20):
    patch_x = random.randint(0, ground_image.get_width() - 50)
    patch_y = random.randint(0, ground_image.get_height() - 20)
    patch_width = random.randint(20, 50)
    patch_height = random.randint(10, 20)
    pygame.draw.ellipse(ground_image, DARK_GREEN, (patch_x, patch_y, patch_width, patch_height))

ground_scroll_x = 0 # Current x position of the ground image for scrolling

# Game clock to control frame rate
clock = pygame.time.Clock()

# Main game loop
while True:
    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE or event.key == pygame.K_UP:
                if GAME_ACTIVE:
                    player.jump()
                else: # If game is over, pressing jump restarts it
                    reset_game()
            if event.key == pygame.K_ESCAPE:
                pygame.quit()
                sys.exit()

    if GAME_ACTIVE:
        # --- Update game elements ---
        player.update()

        # Update and draw obstacles
        obstacles.update()

        # Manage obstacles: remove off-screen, check for clearing for score
        for obstacle in list(obstacles): # Iterate over a copy to safely remove items
            if obstacle.rect.right < 0: # Obstacle is completely off-screen to the left
                obstacles.remove(obstacle)
            # Check if player has successfully cleared an obstacle
            if not obstacle.cleared and obstacle.rect.right < player.rect.left:
                raw_score += 100 # Add 100 to raw_score, which translates to 10 display points
                obstacle.cleared = True

        # Obstacle spawning logic
        OBSTACLE_SPAWN_TIMER -= 1
        if OBSTACLE_SPAWN_TIMER <= 0:
            spawn_obstacle()
            OBSTACLE_SPAWN_TIMER = random.randint(OBSTACLE_SPAWN_INTERVAL_MIN, OBSTACLE_SPAWN_INTERVAL_MAX)

        # Collision detection with obstacles
        for obstacle in obstacles:
            if obstacle.type == "puddle":
                # Puddle collision: Player falls in if bottom is below ground AND x-position aligns with puddle
                if player.rect.bottom > GROUND_LEVEL and \
                   player.rect.left < obstacle.rect.right and \
                   player.rect.right > obstacle.rect.left:
                    GAME_ACTIVE = False
                    play_collision_sound()
            else: # Collision for solid obstacles (walls, gnomes, fowl)
                if player.rect.colliderect(obstacle.rect):
                    GAME_ACTIVE = False
                    play_collision_sound()

        # Score update (time-based survival)
        raw_score += 1 # Increments every frame, displayed as raw_score // 10

        # Ground scrolling update
        ground_scroll_x -= GAME_SPEED
        if ground_scroll_x <= -SCREEN_WIDTH: # When one ground_image has scrolled off, reset position
            ground_scroll_x = 0

        # --- Drawing game elements ---
        screen.fill(BLUE_SKY) # Fill background with sky color

        # Draw clouds with a parallax effect (scrolls slower than foreground)
        cloud_scroll_speed = GAME_SPEED * 0.2
        # Use modulo operator to make clouds loop
        pygame.draw.circle(screen, WHITE, (int((100 + ground_scroll_x * cloud_scroll_speed) % (SCREEN_WIDTH + 200) - 100), 80), 30)
        pygame.draw.circle(screen, WHITE, (int((150 + ground_scroll_x * cloud_scroll_speed) % (SCREEN_WIDTH + 200) - 100), 60), 40)
        pygame.draw.circle(screen, WHITE, (int((200 + ground_scroll_x * cloud_scroll_speed) % (SCREEN_WIDTH + 200) - 100), 70), 35)

        pygame.draw.circle(screen, WHITE, (int((400 + ground_scroll_x * cloud_scroll_speed) % (SCREEN_WIDTH + 200) - 100), 120), 25)
        pygame.draw.circle(screen, WHITE, (int((450 + ground_scroll_x * cloud_scroll_speed) % (SCREEN_WIDTH + 200) - 100), 100), 30)
        pygame.draw.circle(screen, WHITE, (int((500 + ground_scroll_x * cloud_scroll_speed) % (SCREEN_WIDTH + 200) - 100), 110), 28)

        # Draw ground (two instances for continuous scrolling effect)
        screen.blit(ground_image, (ground_scroll_x, GROUND_LEVEL))
        screen.blit(ground_image, (ground_scroll_x + SCREEN_WIDTH, GROUND_LEVEL))

        player.draw(screen) # Draw Sir Hopsalot
        for obstacle in obstacles:
            obstacle.draw(screen) # Draw each obstacle

        # Draw score display
        display_score = raw_score // 10 # Convert raw_score to display score (1 point per 10 frames)
        score_text = FONT.render(f"Score: {display_score}", True, BLACK)
        screen.blit(score_text, (SCREEN_WIDTH - score_text.get_width() - 20, 20))

    else: # Game Over state: display messages
        screen.fill(BLUE_SKY) # Clear screen and show sky background
        game_over_text = LARGE_FONT.render("Game Over!", True, BLACK)
        final_score_text = FONT.render(f"Final Score: {raw_score // 10}", True, BLACK)
        restart_text = FONT.render("Press SPACE to restart", True, BLACK)

        # Center the text on the screen
        screen.blit(game_over_text, (SCREEN_WIDTH // 2 - game_over_text.get_width() // 2, SCREEN_HEIGHT // 2 - 50))
        screen.blit(final_score_text, (SCREEN_WIDTH // 2 - final_score_text.get_width() // 2, SCREEN_HEIGHT // 2 + 10))
        screen.blit(restart_text, (SCREEN_WIDTH // 2 - restart_text.get_width() // 2, SCREEN_HEIGHT // 2 + 50))

    # Update the full display surface to the screen
    pygame.display.flip()
    # Control frame rate
    clock.tick(60) # Limits game to 60 frames per second
````
